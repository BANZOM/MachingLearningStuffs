```python
# finding the percentage of null values

percentage =( train_df.isnull().sum() / train_df.shape[0] ) * 100

percentage 
```

```python
# identifying column if they have more than 17% null data

target_columns = percentage[ percentage > 17].keys()
print(target_columns)

```

```python
# dropping those columns i.e dropping target columns from main dataframe

modified_train_df = train_df.drop( columns= target_columns)
```

```python
# now checking the new shape of new dataframe

modified_train_df.shape
```
